```python
from pyspark import SparkConf
from pyspark import SparkContext
import socket
```

```python
conf=SparkConf().setAppName("161805027_MuhammedÇınaklı")
sc=SparkContext(conf=conf)
```

```python
print(socket.gethostname())
print (socket.gethostbyname(socket.gethostname()))
```

```python
rdd=sc.textFile("161805027_MuhammedÇINAKLI_(27-36).txt")
rdd.count()
```

```python
rdd.take(10)
```

```python
def lower_clean_str(x):
    punc='!"#$%&/()=?-_|\}''][{£><-*/+<>|@,.:;^''+é1234567890'
    lowercased_str = x.lower()
    for ch in punc:
        lowercased_str = lowercased_str.replace(ch,'')
    return lowercased_str
rdd = rdd.map(lower_clean_str)
rdd.take(10)
```

```python
rdd = rdd.flatMap(lambda ln: ln.split(" "))
```

```python
rdd.take(10)
```

```python
rdd = rdd.filter(lambda x: x!='')
```

```python
rdd.take(5)
```

```python
rdd = rdd.filter(lambda x:x!= 'e')
rdd = rdd.filter(lambda x:x!= 'r')
rdd = rdd.filter(lambda x:x!= 'q')
rdd = rdd.filter(lambda x:x!= 'w')
rdd = rdd.filter(lambda x:x!= 'n')
```

```python
rdd.take(5)
```

```python
count=rdd.map(lambda word:(word,1))
```

```python
count.take(5)
```

```python
count_rbk=count.reduceByKey(lambda x,y:(x+y)).sortByKey()
```

```python
count_rbk.take(50)
```

```python
count_rbk = count_rbk.map(lambda x:(x[1],x[0]))
```

```python
count_rbk.take(10)
```

```python
count_rbk.sortByKey(False).take(10)
```

```python
import nltk
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
stopwords
```

```python
count_rbk=count_rbk.filter(lambda x: x[1] not in stopwords).sortByKey(False)
```

```python
count_rbk.sortByKey(False).take(20)
```
